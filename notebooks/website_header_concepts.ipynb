# Website Header Concepts

This notebook turns prior `soma_vs.pkl` runs into wide, hero-style visuals for the lab website.

## Demo Set

1. `Voltage Silk`: layered soma traces with strong negative space on the left. Best static hero if the goal is immediate visual polish.
2. `Gamma Constellation`: spike constellations plus population/LFP bands. Best if you want the header to feel unmistakably scientific.
3. `Population Tides`: smoothed ensemble activity with a few featured traces. Best if the page needs large title text over the art.

## Fast Notes

- Default source run: `results/Old Debug/GammaSignature`
- Preview PNGs are written to `media/website_header_concepts/`
- Swap `RUN_DIR` to try a different run folder that contains `soma_vs.pkl`, `lfp.pkl`, and `input_times.pkl`


```python
from pathlib import Path
from IPython.display import Image, display

from website_header_concepts import (
    load_run,
    plot_gamma_constellation,
    plot_population_tides,
    plot_voltage_silk,
    save_concept_previews,
    summarize_run,
)

RUN_DIR = Path('../results/Old Debug/GammaSignature').resolve()
OUTPUT_DIR = Path('../media/website_header_concepts').resolve()
WINDOW_MS = 650.0

bundle = load_run(RUN_DIR)
summary = summarize_run(bundle)
summary
```
