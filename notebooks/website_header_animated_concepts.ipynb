# Animated Website Header Concepts

This notebook builds animated, sweep-driven hero concepts from the saved `gaba_gmax` sweep instead of repackaging standard analysis plots.

## Why This Sweep

The `gaba_gmax` sweep has a clear activity gradient across 21 completed runs, which makes it a much better animation source than the recent odor-concentration sweep that stayed mostly quiescent.

## Concept Set

1. `aurora_currents`: an evolved, more cinematic version of the population-tides idea.
2. `spike_bloom`: particle / firefly motion driven by spike-density structure.
3. `dendritic_pulse`: real morphology silhouettes with emissive activity fields.
4. `spectral_veil`: raster-field treatment driven by sweep spectrogram structure.


```python
from pathlib import Path
from IPython.display import Image, display

from website_header_animated_concepts import export_all_concepts, load_gaba_sweep_summaries

OUTPUT_DIR = Path('../media/website_header_animated_concepts').resolve()
FRAMES_PER_STEP = 2  # raise to 3 or 4 for smoother loops if you want to spend more render time

summaries = load_gaba_sweep_summaries()
{
    'frames': len(summaries),
    'values': [round(s.value, 2) for s in summaries],
    'output_dir': str(OUTPUT_DIR),
}
```
